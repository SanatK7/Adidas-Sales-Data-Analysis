```python
df = pd.read_excel('/content/Adidas US Sales Datasets.xlsx',skiprows=4)
```

```python
print(df)
```

```python
#Pie Chart for sales by method

sales_by_method = df.groupby('Sales Method')['Total Sales'].sum()

# Plotting a pie chart
plt.figure(figsize=(8, 8))
plt.pie(sales_by_method, labels=sales_by_method.index, autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Sales Method')
plt.show()
```

```python
sales_by_retailer = df.groupby('Retailer')['Total Sales'].sum()

# Plotting a bar chart
plt.figure(figsize=(12, 6))
sales_by_retailer.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Sales by Retailer')
plt.xlabel('Retailer')
plt.ylabel('Total Sales')
plt.show()
```

```python
df['Invoice Date'] = pd.to_datetime(df['Invoice Date'])

# Extract the month from the 'Invoice Date' and create a new 'Month' column
df['Month'] = df['Invoice Date'].dt.month_name()

# Group the data by 'Month' and calculate the total sales for each month
sales_by_month = df.groupby('Month')['Total Sales'].sum()

# Plotting a line graph
plt.figure(figsize=(12, 6))
sales_by_month.plot(kind='line', marker='o', color='blue', linestyle='-', linewidth=2)
plt.title('Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```
